## Converters parameter:

```python
def rename(name):
    if name == 'Royal Challengers Bangalore':
        return 'RCB'
    else:
        return name
```

```python
df = pd.read_csv('IPL Matches 2008-2020.csv', converters= {'team1': rename})
df['team1'].head()
```
